```python
import pandas as pd
import numpy as np

# Đọc dữ liệu
data = pd.read_csv('babies.txt', delim_whitespace=True)

# Phân chia dữ liệu theo giá trị 'smoke'
data_smoke_1 = data[data['smoke'] == 1]
data_smoke_0 = data[data['smoke'] == 0]

# Các thống kê cần tính
stats = ['Std', 'Var', 'Range', 'IQR', 'Skewness', 'Kurtosis']

# Tính các giá trị thống kê cho nhóm hút thuốc
value_smoke_1 = [
    data_smoke_1['bwt'].std(),
    data_smoke_1['bwt'].var(),
    data_smoke_1['bwt'].max() - data_smoke_1['bwt'].min(),  # Range
    data_smoke_1['bwt'].quantile(0.75) - data_smoke_1['bwt'].quantile(0.25),  # IQR
    data_smoke_1['bwt'].skew(),
    data_smoke_1['bwt'].kurtosis()
]

# Tính các giá trị thống kê cho nhóm không hút thuốc
value_smoke_0 = [
    data_smoke_0['bwt'].std(),
    data_smoke_0['bwt'].var(),
    data_smoke_0['bwt'].max() - data_smoke_0['bwt'].min(),  # Range
    data_smoke_0['bwt'].quantile(0.75) - data_smoke_0['bwt'].quantile(0.25),  # IQR
    data_smoke_0['bwt'].skew(),
    data_smoke_0['bwt'].kurtosis()
]

# Chuyển đổi danh sách sang mảng NumPy để thực hiện phép trừ
value_smoke_1 = np.array(value_smoke_1)
value_smoke_0 = np.array(value_smoke_0)

# Tạo DataFrame chứa các thống kê
df_stats = pd.DataFrame({
    'Index': stats,
    'TH1: Bà mẹ hút thuốc': value_smoke_1,
    'TH2: Bà mẹ không hút thuốc': value_smoke_0,
    'Chênh lệch (TH2 - TH1)': value_smoke_0 - value_smoke_1
})

# Hiển thị kết quả
df_stats
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file babies.txt
df = pd.read_csv('babies.txt', delim_whitespace=True)

# Lọc chỉ lấy các giá trị hợp lệ trong cột smoke
df = df[df['smoke'].isin([0, 1])]

# Tạo boxplot
plt.figure(figsize=(10, 6))
box = plt.boxplot(
    [df[df['smoke'] == 1]['bwt'], df[df['smoke'] == 0]['bwt']],
    labels=['Có hút thuốc', 'Không hút thuốc'],
    patch_artist=True  # Cho phép tùy chỉnh màu sắc
)

# Tùy chỉnh màu sắc
colors = ['lightblue', 'lightblue']  # Màu sắc cho các box
for median, color in zip(box['medians'], colors):
    median.set(color='red')  # Màu cho đường median

for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)  # Màu cho box

# Đặt tên cho các trục và tiêu đề
plt.xlabel('')  # Loại bỏ nhãn của trục x
plt.ylabel('Cân nặng (gram)')
plt.title('Boxplot so sánh cân nặng giữa bà mẹ có hút thuốc và không hút thuốc')

# Hiển thị boxplot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('lungcancer.csv')

# Tạo biểu đồ hình tròn
plt.figure(figsize=(10, 6))
labels = ['Không mắc bệnh ung thư', 'Mắc bệnh ung thư']
data = data['Result'].value_counts()

# Thêm màu sắc và tiêu đề
colors = ['#ffcc66', '#9999ff']
#x: Giá trị từng phần
#autopct='%1.1f%%': Hiển thị phần trăm trên biểu đồ, với 1 chữ số thập phân
#startangle=90: Xoay biểu đồ bắt đầu từ góc 90 độ
plt.pie(x=data, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90)

# Thêm tiêu đề
plt.title('Tỷ lệ giữa bệnh nhân mắc bệnh và không mắc bệnh ung thư phổi')
plt.axis('equal') 
plt.show()
```
