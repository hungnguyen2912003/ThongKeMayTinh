```python
#####################################################
# Xây dựng Histogram
#Cách 1:
import pandas as pd
dframe=pd.DataFrame({
    'length' : [1.5,0.5,1.2,0.9,3],
    'width': [0.7,0.2,0.15,0.2,1.1]
}, index=['pig','rabbit','duck','chicken','horse'])
hist=dframe.hist(bins=3)
```

```python
#Cách 2
import matplotlib.pyplot as plt
import numpy as np
d = np.random.uniform(0, 40, 100)
#n: danh sách số lượng phần tử trong mỗi bin
#bins: Mảng chứa các giá trị biên của mỗi bin
#patches: Đối tượng đại diện cho mỗi hình chữ nhật trong biểu đồ
n, bins, patches  = plt.hist(x = d, bins = 'auto', color = '#0504aa', alpha = 0.7, rwidth = 0.85)
#alpha: Độ trong suốt của các cột, 0 < alpha < 1
#rwidth: Độ rộng của các cột trong biểu đồ
#Hiển thị các đường lưới trên trục y (tần suất)
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('My Very Own Histogram')
#Thêm văn bản vào biểu đồ tại toạ độ x = 23, y = 45
#$\mu=15, \sigma=3$ là công thức toán học viết dưới dạng LaTeX (ký hiệu trung bình và độ lệch chuẩn)
plt.text(23, 45, r'$\mu=15, \sigma=3$')
maxfreq=n.max()

#plt.ylim(): Giới hạn của trục y
plt.ylim(ymax=np.ceil(maxfreq/10)*10 if maxfreq % 10 else maxfreq+10)
```

```python
########################################################
#QQ plot
from numpy.random import seed
from numpy.random import randn
from statsmodels.graphics.gofplots import qqplot
from matplotlib import pyplot
#seed the random number generator
seed(1)
#generate univariate observations
data=5*randn(100)+50
#q-q plot
qqplot(data,line='s')
pyplot.show()
```
